```python
import pandas as pd
import numpy as np

def clean_datset(dat):
    """
    Cleans the datset by removing all columns containing NaN or non-standard missing values.

    Parameters:
    - dat (pd.datFrame): The datset to clean.

    Returns:
    - pd.datFrame: Cleaned datset with no columns containing NaN values.
    """
    # Replace non-standard missing values with NaN
    dat = dat.replace([np.inf, -np.inf], np.nan)  # Handle numeric infinities
    dat = dat.replace(r'^\s*$', np.nan, regex=True)  # Replace empty or whitespace-only strings
    dat = dat.replace(['N/A', 'n/a'], np.nan)  # Replace textual missing values

    # Drop columns with any NaN values 

    return dat


clean_datset(data)
```

```python

def find_nonnumeric_entries_in_numeric_columns(dat):
    """
    Identifies and lists all unique non-numeric entries in numeric (int or float) columns of a datset.

    Parameters:
    - dat (pd.datFrame): The datset to analyze.

    Returns:
    - dict: A dictionary where keys are numeric column names and values are sets of non-numeric entries.
    """
    nonnumeric_entries = {}

    # Select only numeric (int and float) columns
    numeric_columns = dat.select_dtypes(include=[np.number]).columns

    for column in numeric_columns:
        # Identify non-numeric entries in the numeric column
        nonnumeric = dat[~dat[column].apply(lambda x: isinstance(x, (int, float, np.integer, np.floating)))]
        unique_nonnumeric = nonnumeric[column].dropna().unique()

        if len(unique_nonnumeric) > 0:
            nonnumeric_entries[column] = set(unique_nonnumeric)

    return nonnumeric_entries

find_nonnumeric_entries_in_numeric_columns(data)
```

```python
import pandas as pd
import numpy as np

def bootstrap_imputation(datset):
    """
    Fills missing values in both numeric and non-numeric columns using bootstrap resampling.
    
    Parameters:
    - datset (pd.datFrame): The datset with missing values.

    Returns:
    - pd.datFrame: datset with missing values filled using bootstrap resampling.
    """
    def bootstrap_fill(column):
        """
        Fills missing values in a single column using bootstrap resampling.
        """
        non_missing = column.dropna()
        
        if len(non_missing) == 0:
            return column
        
        return column.apply(lambda x: np.random.choice(non_missing) if pd.isna(x) else x)
    
    imputed_dat= datset.apply(bootstrap_fill, axis=0)
    return imputed_dat

bootstrap_imputation(data)
```

```python
def inspect_null_types(datset):
    """
    Identifies and categorizes the types of null values in the datset.
    
    Parameters:
    - datset (pd.datrame): The datet to inspect.

    Returns:
    - pd.datFrame: Summary of null value types and their counts for each column.
    """
    null_summary = {}
    null_like_values = [np.nan, None, '', ' ', 'null', 'NULL', 'NaN', 'nan', 'nanns', 'N/A', 'n/a']
    for column in datset.columns:
        column_summary = {}
        for null_value in null_like_values:
            count = datset[column].apply(lambda x: str(x).strip().lower() == str(null_value).strip().lower()                                          if isinstance(x, str) else pd.isna(x)).sum()
            if count > 0:
                column_summary[str(null_value)] = count
        if column_summary:
            null_summary[column] = column_summary

    summary_df = pd.DataFrame.from_dict(null_summary, orient='index').fillna(0).astype(int)
    return summary_df

inspect_null_types(data)
```
