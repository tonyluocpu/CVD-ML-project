## FAISS for scalability

```python
import faiss
import numpy as np
from sentence_transformers import SentenceTransformer

# Extract relevant columns for matching
w = data["statecounty_x"].astype(str).tolist()  # Entries in 'data'
i = final_merged["Code"].astype(str).tolist()  # Entries in 'final_merged'

# Load SBERT model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Function to compute prefix match length
def prefix_match_length(query, candidate):
    match_len = 0
    for q_char, c_char in zip(query, candidate):
        if q_char == c_char:
            match_len += 1
        else:
            break
    return match_len

# Step 1: Prefix Matching
results = []
for query in w:
    prefix_scores = [(candidate, prefix_match_length(query, candidate)) for candidate in i]
    max_prefix_length = max(score for _, score in prefix_scores)
    filtered_candidates = [candidate for candidate, score in prefix_scores if score == max_prefix_length]
    results.append((query, filtered_candidates))

# Step 2: Semantic Search with FAISS
final_results = []
for query, filtered_candidates in results:
    if not filtered_candidates:
        final_results.append((query, None))  # No match
        continue

    # Compute embeddings for filtered candidates and the query
    query_embedding = model.encode([query], convert_to_tensor=False)
    candidate_embeddings = model.encode(filtered_candidates, convert_to_tensor=False)

    # Normalize embeddings for cosine similarity
    query_embedding = np.array(query_embedding / np.linalg.norm(query_embedding, axis=1, keepdims=True))
    candidate_embeddings = np.array(candidate_embeddings / np.linalg.norm(candidate_embeddings, axis=1, keepdims=True))

    # Build FAISS index for filtered candidates
    dimension = query_embedding.shape[1]
    index = faiss.IndexFlatIP(dimension)  # Inner Product Index
    index.add(candidate_embeddings)

    # Perform ANN search
    _, indices = index.search(query_embedding, 1)  # Top 1 match

    # Map results back to candidate names
    best_match_idx = indices[0][0]
    best_match = filtered_candidates[best_match_idx]
    final_results.append((query, best_match))

# Convert results to DataFrame
final_results_df = pd.DataFrame(final_results, columns=["Query", "Best Match"])

# Print unmatched entries and their best semantic match
print("Final Results:")
print(final_results_df)
```
