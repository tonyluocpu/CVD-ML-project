```python
def print_summary_stats(df):
    mean_values = df.mean()
    median_values = df.median()
    range_values = df.max() - df.min()

    summary_stats = pd.DataFrame({
        'Mean': mean_values,
        'Median': median_values,
        'Range': range_values
    })

    print(summary_stats)
print_summary_stats(female_pop_data)
```

```python


def print_summary_stats_nonzero(df):
    df_non_zero = df[df != 0].dropna()
    mean_values = df_non_zero.mean()
    median_values = df_non_zero.median()
    range_values = df_non_zero.max() - df_non_zero.min()

    summary_stats = pd.DataFrame({
        'Mean': mean_values,
        'Median': median_values,
        'Range': range_values
    })

    print(summary_stats)

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

asian_pop_data['Race'] = 'Asian'
black_pop_data['Race'] = 'Black'
aian_pop_data['Race'] = 'AIAN'
nhpi_pop_data['Race'] = 'NHPI'

asian = asian_pop_data.iloc[:, -6:]
black = black_pop_data.iloc[:, -6:]
aian = aian_pop_data.iloc[:, -6:]
nhpi = nhpi_pop_data.iloc[:, -6:]
race_box = pd.concat([asian, black, aian, nhpi])

race_box_long = race_box.melt(
    id_vars=['Race'], 
    var_name='Occupation', 
    value_name='Percentage'
)

race_box_long['Occupation'] = (race_box_long['Occupation']
                               .str.replace('percent_population_in_', '', regex=False)
                               .str.replace('_acs2016_20', '', regex=False)
                               .str.replace('_', ' ')
                               .str.replace('single race', '', regex=False)
                               .str.replace('occupation', '', regex=False)
                               .str.replace('aa', '', regex=False)
                               .str.title()
                               .str.strip())

race_box_long = race_box_long[(~race_box_long['Percentage'].isna()) & (race_box_long['Percentage'] != 0)]

plt.figure(figsize=(16, 8))
sns.boxplot(
    x='Occupation', y='Percentage', hue='Race', data=race_box_long,
    palette='Set2', dodge=False
)

plt.title('Occupational Percentages Across Races (Filtered)', fontsize=16)
plt.xlabel('Occupations', fontsize=12)
plt.ylabel('Percentage', fontsize=12)
plt.xticks(rotation=30, ha='right')
plt.legend(title='Race', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np


df_long = race_box.melt(
    id_vars=['Race'], 
    var_name='Variable',  
    value_name='Value'  
)

df_long['Class'] = df_long['Variable'].str.extract(r'^(.*?)_single_race')  
df_long['Variable'] = df_long['Variable'].str.extract(r'percent_population_in_(.*?)_occupations') 

df_long['Variable'] = df_long['Variable'].str.replace('_', ' ').str.title() 
df_long['Class'] = df_long['Class'].str.replace('_', ' ').str.title() 

df_long = df_long.dropna(subset=['Value'])
df_long = df_long[df_long['Value'] != 0]


# df_long.zero().sum().sum()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.DataFrame(df_long)
df['Class'] = df['Class'].replace('Black Aa', 'Black')
df['Variable'] = df['Variable'].replace(
    'Natural Resources Construction And Maintenance',
    'Natural Resources, Construction And Maintenance'
)
plt.figure(figsize=(12, 6))
colors = ['skyblue', 'salmon', 'lightgreen', 'gold', 'plum']
box = []
x_positions = np.arange(1, 6)
width = 0.15

for i, cls in enumerate(df['Class'].unique()):
    subset = df[df['Class'] == cls]
    box.append(plt.boxplot(
        [
            subset[(subset['Variable'] == var) & (subset['Value'] != 0)]['Value'] 
            for var in df['Variable'].unique()
        ],
        positions=x_positions + (i - 2) * width,
        widths=0.1,
        patch_artist=True,
        boxprops=dict(facecolor=colors[i]),
        medianprops=dict(color='black'),
        flierprops=dict(marker='o', markersize=5)
    ))

plt.xticks(x_positions, df['Variable'].unique(), rotation=30, ha='right', fontsize=10)
plt.xlabel('Occupations')
plt.ylabel('Percentage of Population')
plt.title('Occupational Segregation by Population')
for i, cls in enumerate(df['Class'].unique()):
    if i == 'Black Aa':
        i = 'Black'
    plt.plot([], [], color=colors[i], label=cls)
plt.legend(title="Races", loc='upper left', fontsize=8, title_fontsize=10)
plt.tight_layout()
plt.show()
```
