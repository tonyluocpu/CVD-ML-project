```python
X_std = standarize_dataframe(data_impute, None, 'robust')
#X_std = pd.DataFrame(X_std.select_dtypes(include='number'), columns=X_std.columns)
X_std = X_std.select_dtypes(include='number')
```

```python
X_std.isnull().sum().sum()
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error

X_train, X_test, y_train, y_test = train_test_split(X_std, Y, test_size=0.2, random_state=42)
ridge_model = Ridge(alpha=1.0) 
ridge_model.fit(X_train, y_train)
y_pred = ridge_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
print("Ridge Coefficients:", ridge_model.coef_)
```

```python
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

import numpy as np

threshold = 0.1
significant_features = np.abs(ridge_model.coef_.flatten()) >= threshold

X_train_reduced = X_train.loc[:, significant_features]
X_test_reduced = X_test.loc[:, significant_features]

ridge_model_reduced = Ridge(alpha=1.0)
ridge_model_reduced.fit(X_train_reduced, y_train)
y_pred_reduced = ridge_model_reduced.predict(X_test_reduced)
mse_reduced = mean_squared_error(y_test, y_pred_reduced)

print(f"Reduced Model Mean Squared Error: {mse_reduced}")
print("Reduced Ridge Coefficients:", ridge_model_reduced.coef_)

```

```python

significant_columns = X_train.columns[significant_features].tolist()

print("Significant Columns:")
print(significant_columns)

import pandas as pd
pd.DataFrame(significant_columns, columns=["Significant Features"]).to_csv("significant_columns.csv", index=False)

with open("significant_columns.txt", "w") as f:
    for col in significant_columns:
        f.write(f"{col}\n")

stored_significant_columns = significant_columns
len(stored_significant_columns)
```
