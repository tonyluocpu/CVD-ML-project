```python
import pandas as pd
immigration = pd.read_csv("immigration1.csv")
immigration
```

```python
df = immigration[immigration["name"] != "State Total"]
df
```

```python
df['name'] = df['name'].str.replace(r"\s*County$", "", regex=True)
df = df.sort_values(by='name').reset_index(drop=True)
df
```

```python
df.to_csv('immigration_cleaned.csv', index=False)
```
