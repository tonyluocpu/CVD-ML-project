```python
import pandas as pd

pd.set_option('display.max_rows', 200)
```

```python
wiki = pd.read_csv("wikicounty.csv")
wiki
```

```python
wiki = wiki.sort_values(by = "County [2]")
wiki = wiki.reset_index(drop = True)
wiki
```

```python
data = pd.read_csv("orginal_data.csv")
data
```

```python
type(data["county_name"])
type(wiki["County [2]"])
```

```python
# Define a helper function to extract the first word and convert to lowercase
def extract_first_word(value):
    if isinstance(value, str):  # Check if the value is a string
        return value.split()[0].lower()  # Take the first word and convert to lowercase
    return value  # If not a string, return the value as is

# Apply the function to each entry in 'county_name' in data
data["county_name"] = [extract_first_word(val) for val in data["county_name"]]

# Apply the function to each entry in 'County [2]' in wiki
wiki["County [2]"] = [extract_first_word(val) for val in wiki["County [2]"]]

# Display the first few rows to verify the changes


```

```python
data = data.sort_values(by = "county_name")
data = data.reset_index(drop=True)
data
```

```python
wiki = wiki.sort_values(by = "County [2]")
wiki
```

```python
merged_df = pd.merge(wiki, data, left_on='County [2]', right_on='county_name', how='inner')

# Reorder the columns to ensure 'county_name' and 'County [2]' are the first two columns
columns_order = ['county_name', 'County [2]'] + [col for col in merged_df.columns if col not in ['county_name', 'County [2]']]
merged_df = merged_df[columns_order]
merged_df
```

```python

# Find non-matching rows based on the lowercase versions
non_matching_rows = merged_df[merged_df['county_name'] != merged_df['County [2]']]

# Count the number of non-matching rows
non_matching_count = non_matching_rows.shape[0]

# Output the result
print("Number of non-matching rows:", non_matching_count)

```
